# Problem Names

```python
# import all packages
import pandas as pd #note -pandas library is already installed
import numpy as np # note - numpy library is already installed

#import data files as pandas dataframes
filepath = 'TitanicData.csv'

df = pd.read_csv(filepath)  #telling it to grab the pd file and read it

df.head()
```

## Splitting the Names Column

```python
df[['Surname','TempName1']] = df['Name'].str.split(',', expand=True)
df[['Surname','TempName1']]
```

```python
df[['Prefix','TempName2','TempName3']] = df['TempName1'].str.split('.', expand=True)
df[['Prefix','TempName2','TempName3']]
```

```python
df[['GName','SDGName','TempName4']]=df['TempName2'].str.split('(',expand=True)
df[['GName','SDGName','TempName4']]
```

```python
df[['SDGName','TempName5']] = df['SDGName'].str.split(')',expand=True)
df[['SDGName','TempName5']] 
```

```python
df.to_csv('titanic_check.csv')
```
